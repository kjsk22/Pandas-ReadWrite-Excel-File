```python
df = pd.read_excel('Fish.xlsx', sheet_name = 'Sheet2', header=None)
df
```

```python
def convertcrosslength(cell):
    if cell == 'N\A':
        return 0
    return cell

def convertweight(cell):
    if cell == 'N\A':
        return 5
    elif cell == '##':
        return 10
    return cell

df = pd.read_excel('Fish.xlsx', sheet_name = 'Sheet2', converters={'Cross Length(Cm)':convertcrosslength,
                                                                  'Weight(Grams)': convertweight})
df
```

```python
df.to_excel('newexcel.xlsx', sheet_name = 'Sheet12', index=False, startrow = 2, startcol = 3)
```
